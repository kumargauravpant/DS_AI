```python
tfidf = TfidfVectorizer()
train_idf = tfidf.fit_transform(person.Text)
```

```python
a= train_idf.toarray()
```

```python
kdtree = KDTree(a, leaf_size = 3)
```

```python
personname = input('Enter the name of the person')
```

```python
person['tfidf'] = list(train_idf.toarray())
```

```python
person
```

```python
personname = input('Enter the name of the person')
distance, idx = kdtree.query(person['tfidf'][person['Name']==personname].tolist(), k=3)
for i, value in list(enumerate(idx[0])):
    print(f"Name: {person['Name'][value]}")
    print(f"Distaince: {distance[0][i]}")
    print(f"URI: {person['URI'][i]}")
```
