```python
# Loading the model 
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
print(doc)
```

```python
type(doc)
```

```python
# Sentense Token
sentense = nlp(text)

# Identify the element
sentense = list(sentense.sents)
print(sentense)
print(len(sentense))
```

```python
for sent in sentense:
    print(sent, '\n')
```

```python
# Word Token
word = nlp(text1)
# tokenize
for token in word:
    print(token)
```

```python
# Check the stop words for english words
stopwords = spacy.lang.en.stop_words.STOP_WORDS
print(stopwords)
```

```python
print(len(stopwords))
```

```python
doc = nlp(text1)
# tokenize
for token in doc:
    if token.is_stop:
        print(token)
```

```python
#Printing sentense without token
doc = nlp(text1)
for token in doc:
    if not token.is_stop:
        print(token)
```

```python
#Punctuation
doc = nlp(text1)
for token in doc:
    if not token.is_punct:
        print(token)
```

```python
doc = nlp(text)
for token in doc:
    if not token.is_punct and not token.is_stop:
        print(token)
```

```python
' '.join([token.text for token in doc if not token.is_punct and not token.is_stop])
```

```python
#lemma or stemming
doc = nlp(text1)

for token in doc:
    print(f'{token} -->{token.lemma_}')
    
```

```python
for token in doc:
    print(f'{token} , ascii --> {token.is_ascii}, lower --> {token.is_lower}')
```

```python
#part of speech
doc = nlp(text)

for token in doc:
    print(f'{token} -->{token.tag_}')
```

```python
#part of speech
doc = nlp(text)

for token in doc:
    print(f'{token} -->{token.pos_} {spacy.explain(token.tag_)}')
```

```python
# Dependency Parsing
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True, options={'distance':100})
```

```python
#NER
doc = nlp(text)

for token in doc.ents:
    print(f'{token} -->{token.label_}')
```

```python
from spacy import displacy
displacy.render(doc, style='ent', jupyter=True, options={'distance':100})
```
