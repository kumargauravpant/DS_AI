```python
# Import Stanza
import stanza
```

```python
# Download the language package
stanza.download('en')
```

```python
# tokenize
#abc = stanza.Pipeline('en', processors='tokenize')
#abc = stanza.Pipeline('en', processors='tokenize,lemma')
abc = stanza.Pipeline('en')
```

```python
doc = abc('Barack Obama was born in hawaii, I was born in india')
```

```python
type(doc)
```

```python
print(doc)
```

```python
for i in doc.sentences:
    for j in i.tokens:
        print(j.text)
    
```

```python
# POS Tagger
abc = stanza.Pipeline('en', processors='tokenize,pos', use_gpu=True) #use_gpu is to use gpu if availble
```

```python
doc = abc('Barack Obama was born in hawaii, I was born in india')
```

```python
doc
```

```python
for i , sentence in enumerate(doc.sentences):
    print(f'==============Sentence{i+1} tokens ================')
    print(*[f'id: {token.id}\ttext: {token.text}\tupos:{token.upos}'for token in sentence.words] , sep='\n')
```

```python
#Lemma
abc = stanza.Pipeline('en', processors='tokenize,pos,lemma')
```

```python
doc = abc('Barack Obama was born in hawaii, I was born in india')
doc
```

```python
from spacy_stanza import StanzaLanguage
from spacy import displacy
```

```python
abc = stanza.Pipeline('en')
```

```python
# Wrap Stanza with spacy
nlp = StanzaLanguage(abc)
```

```python
doc = nlp('Barack Obama was born in hawaii, I was born in india')
```

```python
displacy.render(doc, style='dep')
```
