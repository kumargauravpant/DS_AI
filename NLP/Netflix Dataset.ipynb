```python
import numpy as np
import pandas as pd
import string
```

```python
from nltk.corpus import stopwords
import re
from sklearn.model_selection import train_test_split
```

```python
pos_rev = pd.read_csv('Data/pos.txt', sep="\n", encoding='latin-1', header=None)
pos_rev['Mood']=1
pos_rev=pos_rev.rename(columns={0:'Review'})
pos_rev
```

```python
neg_rev = pd.read_csv('Data/negative.txt', sep="\n", encoding='latin-1', header=None)
neg_rev['Mood']=0
neg_rev=neg_rev.rename(columns={0:'Review'})
neg_rev
```

```python
# Text Analytics
# 1. Lower Case
# 2. Remove Punctuation
# 3. Remove stopwords
# 4. Concetinate the dataset
# 5. Remove Numbers
# 6. Remove Special Characters
# 7. Remove Links
```

```python
# Pos data
pos_rev.loc[:, 'Review'] = pos_rev.loc[:, 'Review'].apply(lambda x:x.lower())
pos_rev.loc[: , 'Review'] = pos_rev.loc[: , 'Review'].apply(lambda x :x.translate(str.maketrans(dict.fromkeys(string.punctuation))))
pos_rev.loc[: , 'Review'] = pos_rev.loc[: , 'Review'].apply(lambda x :' '.join([i for i in x.split() if i not in stopwords.words('english')]))
pos_rev.loc[: , 'Review'] = pos_rev.loc[: , 'Review'].apply(lambda x : re.sub(r'\d', '', x))
pos_rev
```

```python
# Negative data
# Pos data
neg_rev.loc[: , 'Review'] = neg_rev.loc[:, 'Review'].apply(lambda x:x.lower())
neg_rev.loc[: , 'Review'] = neg_rev.loc[: , 'Review'].apply(lambda x :x.translate(str.maketrans(dict.fromkeys(string.punctuation))))
neg_rev.loc[: , 'Review'] = neg_rev.loc[: , 'Review'].apply(lambda x :' '.join([i for i in x.split() if i not in stopwords.words('english')]))
neg_rev.loc[: , 'Review'] = neg_rev.loc[: , 'Review'].apply(lambda x : re.sub(r'\d', '', x))
neg_rev
```

```python
# Containating the dataset
com_rev = pd.concat([pos_rev, neg_rev], axis=0).reset_index()
com_rev
```

```python
# Train Test Split
X_train, X_test, Y_train , Y_test = train_test_split(com_rev['Review'].values , com_rev['Mood'].values , test_size = 0.2 , random_state = 101)
```

```python
train_data = pd.DataFrame({'review':X_train, 'mood':Y_train})
test_data = pd.DataFrame({'review':X_test, 'mood':Y_test})
```

```python
# Vector Form
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

train_vectors = vectorizer.fit_transform(train_data['review'])
test_vectors = vectorizer.transform(test_data['review'])
```

```python
len(vectorizer.get_feature_names())
```

```python
# SVM
from sklearn import svm
from sklearn.metrics import classification_report
```

```python
classifier_linear = svm.SVC(kernel = 'linear')
classifier_linear.fit(train_vectors, train_data['mood'])
```

```python
prediction_linear = classifier_linear.predict(test_vectors)
```

```python
report = classification_report(test_data['mood'], prediction_linear, output_dict=True)
#print(report)
print('positive', report['1']['recall'])
print('negative', report['0']['recall'])
```
